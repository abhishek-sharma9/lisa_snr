```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
import h5py
```

```python
PSD = np.loadtxt('LISA_psd.txt')
```

```python
plt.figure(figsize = (8, 6))

plt.loglog(PSD[:, 0], PSD[:, 1], c = 'b')

plt.xlabel('frequency (Hz)', fontsize = 10)
# plt.ylabel(r'ASD ($\frac{1}{\sqrt{Hz}}$)', fontsize = 10)
plt.ylabel(r'Power Spectral Density ($1/Hz$)', fontsize = 10)
plt.grid(which = 'both')
# plt.legend(fontsize = 10)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
# plt.savefig('asd_comparison.jpg', dpi = 300)
plt.show()
```

```python
with h5py.File('LISA_SNR.hdf', 'r') as f:
    SNR = f['d=6500Mpc_snr'][()]
```

```python
psi = np.pi/6
observationTime = np.linspace(0, 1, 365)

lamda = np.linspace(-np.pi, np.pi, 360)      #longitude in SSB frame.
beta = np.linspace(-np.pi/2, np.pi/2, 360)   #latitude in SSB frame.

Lamda, Beta = np.meshgrid(lamda, beta)

Beta.shape, Lamda.shape
```

```python
SNR.shape
```

```python
print(SNR.min(), SNR.max())
```

```python
fig, axes = plt.subplots(5, 3, figsize = (20, 20), subplot_kw=dict(projection="mollweide"))

ax1, ax2, ax3 = axes[0, 0], axes[0, 1], axes[0, 2]
ax4, ax5, ax6 = axes[1, 0], axes[1, 1], axes[1, 2]
ax7, ax8, ax9 = axes[2, 0], axes[2, 1], axes[2, 2]
ax10, ax11, ax12 = axes[3, 0], axes[3, 1], axes[3, 2]
ax13, ax14 = axes[4, 0], axes[4, 1]

vmin, vmax = SNR.min(), SNR.max()

im1 = ax1.pcolormesh(Lamda, Beta, SNR[0], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im1, label = 'SNR')
ax1.grid()
ax1.set(title = r'$time = 0$ days')

im2 = ax2.pcolormesh(Lamda, Beta, SNR[1], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im2, label = 'SNR')
ax2.grid()
ax2.set(title = r'$time = 30$ days')

im3 = ax3.pcolormesh(Lamda, Beta, SNR[2], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im3, label = 'SNR')
ax3.grid()
ax3.set(title = r'$time = 60$ days')

im4 = ax4.pcolormesh(Lamda, Beta, SNR[3], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im4, label = 'SNR')
ax4.grid()
ax4.set(title = r'$time = 90$ days')

im5 = ax5.pcolormesh(Lamda, Beta, SNR[4], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im5, label = 'SNR')
ax5.grid()
ax5.set(title = r'$time = 120$ days')

im6 = ax6.pcolormesh(Lamda, Beta, SNR[5], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im6, label = 'SNR')
ax6.grid()
ax6.set(title = r'$time = 150$ days')

im7 = ax7.pcolormesh(Lamda, Beta, SNR[6], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im7, label = 'SNR')
ax7.grid()
ax7.set(title = r'$time = 180$ days')

im8 = ax8.pcolormesh(Lamda, Beta, SNR[7], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im8, label = 'SNR')
ax8.grid()
ax8.set(title = r'$time = 210$ days')

im9 = ax9.pcolormesh(Lamda, Beta, SNR[8], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im9, label = 'SNR')
ax9.grid()
ax9.set(title = r'$time = 240$ days')

im10 = ax10.pcolormesh(Lamda, Beta, SNR[9], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im10, label = 'SNR')
ax10.grid()
ax10.set(title = r'$time = 270$ days')

im11 = ax11.pcolormesh(Lamda, Beta, SNR[10], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im11, label = 'SNR')
ax11.grid()
ax11.set(title = r'$time = 300$ days')

im12 = ax12.pcolormesh(Lamda, Beta, SNR[11], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im12, label = 'SNR')
ax12.grid()
ax12.set(title = r'$time = 330$ days')

im13 = ax13.pcolormesh(Lamda, Beta, SNR[12], cmap=plt.cm.jet, vmin=vmin, vmax=vmax)
fig.colorbar(im13, label = 'SNR')
ax13.grid()
ax13.set(title = r'$time = 360$ days')

fig.delaxes(axes[4, 1])
fig.delaxes(axes[4, 2])

plt.show()
```
